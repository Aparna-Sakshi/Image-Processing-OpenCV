```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image=cv2.imread("flower.jfif")
```

```python
plt.figure(figsize=(20,10))
plt.imshow(image)
```

```python
img=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(20,8))
plt.imshow(img)
```

```python
final_img=cv2.cvtColor(img,cv2.COLOR_RGB2HSV)
```

```python
lower=np.array([24,100,100])
upper=np.array([34,255,255])
```

```python
area_of_interest=cv2.inRange(final_img,lower,upper)
```

```python
plt.figure(figsize=(20,10))
plt.imshow(area_of_interest)
```

```python
result=cv2.bitwise_and(img,img,mask=area_of_interest)
plt.figure(figsize=(20,10))
plt.imshow(result)
```
