```python
#importing libraries
import cv2
import numpy as np
from google.colab.patches import cv2_imshow


img = cv2.imread('tele.jpg')

cv2_imshow(img)


gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2_imshow(gray)


img = cv2.GaussianBlur(gray,(3,3),0)


laplacian = cv2.Laplacian(img,cv2.CV_64F)


cv2_imshow(laplacian)
```
